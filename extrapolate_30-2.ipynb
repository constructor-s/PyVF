```python
def mark(cmatrix, i, j, val):
    if np.isnan(i) or np.isnan(j):
        return

    cmatrix[int(i), int(j)] = val

cmatrix = np.zeros([76, 76], dtype=np.int64) * -1
for i, row in enumerate(P30D2_MATRIX):
    for j, idx in enumerate(row):
        if np.isnan(idx):
            continue
        # Find 8 neighbors
        for dy in (-1, 0, 1):
            for dx in (-1, 0, 1):
                if dx == 0 and dy == 0:
                    continue
                if i == 0 and dy == -1:
                    continue
                if i == P30D2_MATRIX.shape[0]-1 and dy == 1:
                    continue
                if j == 0 and dx == -1:
                    continue
                if j == P30D2_MATRIX.shape[1]-1 and dx == 1:
                    continue
                mark(cmatrix, idx, P30D2_MATRIX[i+dy, j+dx], dx*1+dy*10)
        
plt.imshow(cmatrix)
```
